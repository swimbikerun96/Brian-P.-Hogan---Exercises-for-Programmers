```python
#Constraints

#Handle Input for Pulse
#Error Message display if input invalid
while True:
    try:
        pulse_in = input('Please Enter Resting Pulse: ')
        pulse = int(pulse_in)
        break
    except:
        print('Integer Inputs Only!')
        continue

#Handle Input for Age
#Error Message display if input invalid
while True:
    try:
        age_in = input('Please Enter Age: ')
        age = int(age_in)
        break
    except:
        print('Integer Inputs Only!')
        continue

#Define Function for Karvonen Heart Rate
def karvonen(pulse, age, intensity):
    p = pulse
    a = age
    i = intensity/100
    thr = (((220 - a) - p) * i) + p
    return int(thr)

#Print Elements for Tabular Format
print('')
print('Intensity |  Rate')
print('-------------------')

#Loop to generate data table
#Use loop to generate the percentages
for i in range(0,100,5):
    x = karvonen(pulse, age, i)
    if len(str(i)) == 1:
        print(f'{i}%        |  {x} bpm')
    elif len(str(x)) == 2:
        print(f'{i}%       |  {x} bpm')
    else:
        print(f'{i}%       | {x} bpm')
```

```python
#Challenge

#Import Tkinter Components
import tkinter
from tkinter import *

#Build Window
root = tkinter.Tk()

#Define Karvonen Function
def karvonen(event):
    try:
        a = int(age.get(1.0, END))
        p = int(pulse.get(1.0, END))
        i = w.get()/100
        thr = round((((220 - a) - p) * i) + p,2)
        if w.get() == 0:
            pass
        else:
            label3.config(text = f'{thr} bpm')
    except:
        pass

#Build & Pack the title
label1 = Label(root, text="Karvonen Heart Rate!")
label1.pack()

#Age Widgets Packed & Bound
al = Label(root, text = "Age: ")
al.pack()
age = Text(root, height=1, width= 3)
age.pack()
age.bind("<Key>",karvonen)

#Pulse Widgets Packed & Bound
pl = Label(root, text = "Resting Pulse: ")
pl.pack()
pulse = Text(root, height=1, width= 3)
pulse.pack()
pulse.bind("<Key>",karvonen)

#Intensity Widget Packed & Bound
label2 = Label(root, text="Use the slider to input your level of intensity:")
label2.pack()
#Build the slider
w = Scale(root, from_=0, to=100, orient=HORIZONTAL)
w.pack()
w.bind("<ButtonRelease-1>", karvonen)

#Build Label for output
label3 = Label(root, text="")
label3.pack()

#Run Main Loop
mainloop()
```
